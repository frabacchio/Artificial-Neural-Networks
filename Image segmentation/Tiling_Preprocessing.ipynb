```
# we use the tiling technique without overlapping
# we create a new dataset of images with size 256*256*3
# so every image and mask of the Bipbip dataset is decomposed in 48 images or masks

# Bipbip MAIS DATASET

# we create two folders for images and masks 
!mkdir -p '/content/drive/My Drive/AN2DL/Challenge2/Bip/Training256/Mais/Images/img'
!mkdir -p '/content/drive/My Drive/AN2DL/Challenge2/Bip/Training256/Mais/Masks/img'

from PIL import Image

#function to create and save subimages
def imgcrop(input, xPieces, yPieces,save_path,img_name):
    filename, file_extension = os.path.splitext(input)
    im = Image.open(input).convert('RGB')
    imgwidth, imgheight = im.size
    height = imgheight // yPieces
    width = imgwidth // xPieces
    for i in range(0, yPieces):
        for j in range(0, xPieces):
            box = (j * width, i * height, (j + 1) * width, (i + 1) * height)
            a = im.crop(box)
            a.save(save_path + img_name[:-4] + "-" + str(i) + "-" + str(j) + file_extension)


mais_images_path = '/content/drive/My Drive/AN2DL/Challenge2/Development_Dataset/Training/Bipbip/Mais/Images'
image_filenames = next(os.walk(mais_images_path))[2]   
save_img_path='/content/drive/My Drive/AN2DL/Challenge2/Bip/Training256/Mais/Images/img/'#folder created above
for img_name in image_filenames:
    imgcrop(haricot_images_path+'/'+img_name, 8, 6,save_img_path,img_name)


mais_masks_path = '/content/drive/My Drive/AN2DL/Challenge2/Development_Dataset/Training/Bipbip/Mais/Masks'
image_filenames = next(os.walk(mais_masks_path))[2]   
save_masks_path='/content/drive/My Drive/AN2DL/Challenge2/Bip/Training256/Mais/Masks/img/'#folder created above
for img_name in image_filenames:
   imgcrop(haricot_masks_path+'/'+img_name, 8, 6,save_masks_path,img_name)
```
